```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.random.rand(50,1)
y = 2*X + 1 + 0.1*(np.random.rand(50,1))
```

```python
print(y)
```

```python
plt.plot(X,y,'.')
```

```python
w_new = np.random.rand()
b_new = np.random.rand()
y_pred = w_new * X + b_new
```

```python
plt.plot(X,y,'.')
plt.plot(X,y_pred)
```

```python
lr = 0.1
errors = []
for i in range(100):
  ypred = w_new*X + b_new

  error = (y - ypred)
  errors.extend(error)

  db_new = - 2 * error.mean()
  dw_new = - 2 * (X*error).mean()

  b_new = b_new - lr * db_new
  w_new = w_new - lr * dw_new

  if i % 5 == 0:
    plt.figure(figsize=(10,8))
    plt.plot(X,y,'.')
    plt.plot(X, w_new*X + b_new)
    plt.show()
```
